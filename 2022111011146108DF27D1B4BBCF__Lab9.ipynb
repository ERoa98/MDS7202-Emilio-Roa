## 1.2 Creación de Dataset

Tomando en cuenta la estructura de datos desarrollada en el punto 1.1, construya la clase `MedicalDataset()` que cumpla los siguientes puntos:

- [ ] Poseer un `__init__` en el que se almacene `estructura` creada en 1.1, la `raiz` de la carpeta y una función que permita transformar el dataset (de esto no se preocupe mucho, ya que solamente debe almacenar una función en el atributo).
- [ ] La clase debe ser capaz de entregar la cantidad de elementos a traves de `__len__`.
- [ ] Debe poseer el método `__getitem__` que retorne una tupla con la imagen y su correspondiente etiqueta.

```python
# Código Aquí
class MedicalDataset(Dataset):
    def __init__(self, df, raiz, transform):
        self.df = df
        self.raiz = raiz
        self.transform = transform
    
    def __getitem__(self, idx):
        s = self.df.iloc[idx]
        img_path = s.loc['img_paths']
        
        # Un poco de ayuda para cargar la imagen
        image = Image.open(img_path).convert('RGB')
        label = s.loc['labels']
        
        # Auida para realizar la transformación
        if self.transform:
            image = self.transform(image)
        
        return image, label
    
    def __len__(self):
        return len(self.df)
```

**Respuesta:**

El objetivo del `forward` en este tipo de redes neuronales es calcular el output de la red con las valores de parámetros actuales haciendo todas las multiplicaciones matriciales correspondientes a la red, para así con este output poder obtener un error al comparar el output obtenido con el output deseado.

```python
class FCClassifier(nn.Module):
    def __init__(self, in_channels, num_classes):
        super(FCClassifier, self).__init__()
        self.flatten = nn.Flatten()
        self.lin1 = nn.Linear(in_channels, 16)
        self.lin2 = nn.Linear(16, num_classes)
        
    def forward(self, x):
        batch_size = x.shape[0]
        x = x.view(batch_size, -1)
        h_1 = F.relu(self.lin1(x))
        y_pred = self.lin2(h_1)
    
        return y_pred
```

```python
class CNN_Net(nn.Module):
    
    def __init__(self):
        super(CNN_Net, self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3, padding=1), #128*128*3 -> 128x128x8
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2) #64x64x8
        )
        
        self.fc1 = nn.Linear(in_features=64*64*8, out_features=100)
        self.drop = nn.Dropout2d(0.25)
        self.fc2 = nn.Linear(in_features=100, out_features=50)
        self.fc3 = nn.Linear(in_features=50, out_features=6)
        
    def forward(self, x):
        out = self.layer1(x)
        out = out.view(out.size(0), -1)
        out = self.fc1(out)
        out = self.drop(out)
        out = self.fc2(out)
        out = self.fc3(out)
        
        return out

class CNN_Net2(nn.Module):
    
    def __init__(self):
        super(CNN_Net2, self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3, padding=1), #128*128*3 -> 128x128x8
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2) #64x64x8
        )
        
        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3), #62x62x16
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2) #31x31x16
        )
        
        self.fc1 = nn.Linear(in_features=16*31*31, out_features=100)
        self.drop = nn.Dropout2d(0.25)
        self.fc2 = nn.Linear(in_features=100, out_features=50)
        self.fc3 = nn.Linear(in_features=50, out_features=6)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.view(out.size(0), -1)
        out = self.fc1(out)
        out = self.drop(out)
        out = self.fc2(out)
        out = self.fc3(out)
        
        return out


class CNNClassifier1(nn.Module):
    def __init__(self, in_channels, num_classes):
        super(CNNClassifier1, self).__init__()
        self.conv1 = nn.Conv2d(in_channels , 4, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64*64*4, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16,num_classes)
    def forward(self, x):
        # x es 128 x 128
        x = F.relu(F.max_pool2d(self.conv1(x), 2)) # conv: 128x128x4 -> maxpool: 64x64x4
        x = x.view(-1, 64*64*4)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

    
class CNNClassifier2(nn.Module):
    def __init__(self, in_channels, num_classes):
        super(CNNClassifier2, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, 16, kernel_size=3, padding=1) 
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3)
        self.fc1 = nn.Linear(30*30*64, 64)
        self.fc2 = nn.Linear(64, num_classes)
    def forward(self, x):
        # x es 128 x 128
        x = F.relu(F.max_pool2d(self.conv1(x), 2))# conv: 128x128x16 -> maxpool: 64x64x16
        x = F.relu(F.max_pool2d(self.conv2(x), 2))# conv: 64x64x32 -> maxpool: 32x32x32
        x = F.relu(self.conv3(x))# conv: 30x30x64
        x = x.view(-1, 30*30*64)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
        #return F.log_softmax(x)
```

**Respuesta:**

En la etapa de backward se realiza el proceso de backpropagation, en donde en base al error encontrado entre el output obtenido y el deseado se propaga el error hacia atrás calculando las derivadas y cálculos respectivos para de esta forma saber que parámetros o que partes de la red están más relacionados con el error.

En base a esto posteriormente se puede aplicar el el algoritmo de gradiente descendiente en donde se actualizan los pesos de la red para reducir el error en base a los valores de los gradientes previamente calculados.

```python
# train
def train(model, train_loader, criterion, optimizer, n_epochs):
    since = time.time()

    train_loss = []
    train_accu = []
    
    for e in range(n_epochs):
        model.train()
        running_loss = 0.0
        running_corrects = 0.0
        for i, (data, label) in enumerate(train_loader):
            data = data
            targets = label
            
            #Forward
            optimizer.zero_grad()
            outputs = model(data)
            #print('output',outputs)
            #print('label',label)
            _, preds = torch.max(outputs, 1)
            
            #Backward
            loss = criterion(outputs, label)
            #loss = F.nll_loss(outputs, label)
            loss.backward()
            
            # Descenso del gradiente
            optimizer.step()
            running_loss += loss.item() * data.size(0)
            running_corrects += torch.sum(preds == label.data)

        epoch_loss = running_loss /len(train_dataset)
        epoch_acc = running_corrects.double() / len(train_dataset)

        train_loss.append(epoch_loss)
        train_accu.append(epoch_acc)

        print('Train Loss: {:.4f} Acc: {:.4f}'.format(epoch_loss, epoch_acc))

    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(time_elapsed//60, time_elapsed % 60))

    return model, train_loss, train_accu

def calculate_accuracy(y_pred, y):
  top_pred = y_pred.argmax(1, keepdim=True)
  correct = top_pred.eq(y.view_as(top_pred)).sum()
  acc = correct.float()/y.shape[0]
  return acc

# Evaluate
def evaluate(loader, model, criterion, device):
    epoch_loss = 0
    epoch_acc = 0

    model.eval()
    with torch.no_grad():
        for (x, y) in loader:
            x = x.to(device)
            y = y.long().to(device)

            y_pred = model(x)

            loss = criterion(y_pred, y)

            acc = calculate_accuracy(y_pred, y)

            epoch_loss += loss.item()
            epoch_acc += acc.item()
    return epoch_loss / len(loader), epoch_acc / len(loader)
```
